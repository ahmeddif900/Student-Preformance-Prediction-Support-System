### data_overview

```python
#load both datasets
df_math = pd.read_csv("data/student-mat.csv", sep = ";", index_col=False)
df_por = pd.read_csv("data/student-por.csv", sep = ";", index_col=False)
```

```python
# how many student in math and Portuguese Language
print(f"in math: \n{df_math.shape}")
print(f"in Portuguese Language: \n{df_por.shape}")
```

**i well concatenate two datasets into one dataset**
---
to train one model that uses both groups of students. The "subject" column will help the model know which subject the score came from.

#### ‚úÖ Good for a generalized risk prediction model.
#### ‚ùå Less precise than train two models for each subject, since you lose per-subject granularity.

```python
#make new column (subject)
df_math["subject"] = "math"
df_por["subject"] = "portuguese"
```

```python
#combaine two dataset together
df = pd.concat([df_math,df_por])
#reindexing 
df.index = range(len(df))
df
```

```python
#check for nulls
df.isna().sum()
```

```python
#check for duplicate
df.duplicated().sum()
```

```python
df["G3"]
```

```python
df["risk_level"] = df["G3"].map(lambda x : "high_risk"if x <10 else  "medium_risk" if x<14 else "low_risk")
df["risk_level"]
```

```python
df["risk_level"].value_counts()
```

```python
har_count = df["risk_level"].value_counts()

plt.figure(figsize=(8, 6))
colors = ['#FFD700', '#32CD32', '#FF6B6B']  # Gold, LimeGreen, Light Red
explode = (0.05, 0.05, 0.1)  # Slightly separate slices, emphasize highest risk
# Create the pie chart 
wedges, texts, autotexts = plt.pie(
    har_count.values, 
    labels=har_count.index,
    colors=colors,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=90,
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    pctdistance=0.85
)


for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(11)
    autotext.set_weight('bold')


plt.title('Risk Level Distribution', 
          fontsize=16, 
          fontweight='bold', 
          pad=20)


circle = plt.Circle((0,0), 0.7, color='white', alpha=0.3)
fig = plt.gcf()
fig.gca().add_artist(circle)


plt.axis('equal')

# Add a legend 
plt.legend(wedges, [f'{label}: {count} items' for label, count in zip(har_count.index, har_count.values)],
          title="Risk Levels",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()

plt.show()
```

```python
binary_map = {
    "school": {"GP": 0, "MS": 1},
    "sex": {"F": 0, "M": 1},
    "address": {"U": 0, "R": 1},
    "famsize": {"LE3": 0, "GT3": 1},
    "Pstatus": {"T": 0, "A": 1},
    "schoolsup": {"no": 0, "yes": 1},
    "famsup": {"no": 0, "yes": 1},
    "paid": {"no": 0, "yes": 1},
    "activities": {"no": 0, "yes": 1},
    "nursery": {"no": 0, "yes": 1},
    "higher": {"no": 0, "yes": 1},
    "internet": {"no": 0, "yes": 1},
    "romantic": {"no": 0, "yes": 1},
    "risk_level":{"low_risk":0,"medium_risk":1,"high_risk":2}
}

# Apply binary encoding
df = df.replace(binary_map)

# One-hot encoding for nominal features to avoidnes order
df_encoded = pd.get_dummies(df, columns=["Mjob", "Fjob", "reason", "guardian","subject"], drop_first=True)

#to maka sure all data numric dont have boolean
df_encoded = df_encoded.astype(int)

# Now df is fully numeric
df_encoded.head()
```

```python
# Compute correlation
corr = df_encoded.corr()

# Mask upper triangle
mask = np.triu(np.ones(corr.shape), k=1).astype(bool)

# Check correlations
plt.figure(figsize=(22, 14))
sns.heatmap(corr, mask=mask, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# build subplots to check if the samples are normally distributed
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
plt.suptitle("is scores normally distributed ?")
sns.histplot(df_encoded["G1"],bins=20,kde=True, ax=ax[0])
sns.histplot(df_encoded["G2"],bins=20, kde=True, ax=ax[1])
sns.histplot(df_encoded["G3"],bins=20, kde=True, ax=ax[2])
plt.show()
```

```python
# Separate features and target
X = df_encoded.drop(["risk_level","G3"], axis=1)  # risk level is the target
y = df_encoded["risk_level"]
```

```python
# Model
model = LinearRegression()

# backward Selection
sfs = SFS(model,
          k_features="best",
          floating=False,
          scoring="r2",
          cv=5)

sfs = sfs.fit(X, y)
backward_features = list(sfs.k_feature_names_)
print("Selected Features:", backward_features)
```

```python
# Train model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# Feature importance
importances = pd.Series(rf.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=False)
embedded_features = list(importances[importances > 0.01].index)  # keep >1% importance
print("Embedded Selected:", embedded_features)

# Plot
importances.head(10).plot(kind="barh", figsize=(8,6))
plt.title("Top 10 features by importance")
plt.show()
```

### Cross-Entropy Loss (Log Loss) for Classification

Cross-Entropy Loss measures the difference between the **true labels** and the **predicted probabilities**.  
It is widely used in classification problems because it penalizes incorrect predictions more when the model is confident but wrong.

---

#### 1. Binary Classification

For two classes, where the true label is \( y \in \{0,1\} \) and the predicted probability of class 1 is \( \hat{y} \):

$$
L(y, \hat{y}) = - \big[ y \log(\hat{y}) + (1-y)\log(1-\hat{y}) \big]
$$

$$
If ( y = 1 ), the loss is ( -\log(\hat{y}) ).
$$

$$
If ( y = 0 ), the loss is ( -\log(1 - \hat{y}) ).
$$

---

#### 2. Multiclass Classification

For \( C \) classes, with one-hot true label vector \( y \) and predicted probability vector(≈∑):

$$
L(y, \hat{y}) = - \sum_{c=1}^{C} y_c \cdot \log(\hat{y}_c)
$$

Only the probability of the correct class contributes to the loss.

---

#### 3. Example

Suppose the true label is **cat** (class 1) out of 3 classes:

- True distribution: \([1, 0, 0]\)  
- Predicted distribution: \([0.7, 0.2, 0.1]\)  

$$
L = -\log(0.7) \approx 0.357
$$

If the prediction was \([0.2, 0.7, 0.1]\), then:

$$
L = -\log(0.2) \approx 1.609 \quad (\text{much worse})
$$

---

#### 4. Why Cross-Entropy?

- It encourages the model to output **high probability for the correct class**.  
- Strong penalty for being **confident but wrong**.  
- Standard loss function for **Logistic Regression, Neural Networks, Deep Learning**.

```python
# Train & Evaluate

results = []

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Evaluate with metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)
    
    # Save results
    results.append({
        "Model": name,
        "Accuracy": round(acc*100, 4),
        "Precision": round(prec*100, 4),
        "Recall": round(rec*100, 4),
        "F1-Score": round(f1*100, 4)
    })
    
# Results as DataFrame
results_df = pd.DataFrame(results).sort_values(by="F1-Score", ascending=False)

print("=== Model Performance Comparison ===")
print(results_df)
```

### Deep Learning 

```python
# Build Neural Network
model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),  # Input layer
    Dropout(0.3),                                                  # Dropout to prevent overfitting
    Dense(32, activation="relu"),                                  # Hidden layer
    Dense(len(set(y)), activation="softmax")                       # Output layer (multi-class)
])

# Compile the model
model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy",  # works if y is integer labels
    metrics=["accuracy"]
)

# Train the model
history = model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test),
    epochs=30, 
    batch_size=32,
    verbose=1
)

# Evaluate
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {acc:.4f}")

# Classification report
y_pred = model.predict(X_test).argmax(axis=1)
print(classification_report(y_test, y_pred))
```

```python
# custom cleaner using NLTK
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = re.sub(r"[^a-zA-Z]", " ", text)  # remove non-letters
    tokens = text.lower().split()
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

df["cleaned"] = df["Review"].apply(clean_text)

# train/test split
X_train, X_test, y_train, y_test = train_test_split(df["cleaned"], df["risk_level"], test_size=0.2, random_state=42)

# pipeline with TF-IDF + Logistic Regression
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression(max_iter=1000))
])

pipeline.fit(X_train, y_train)
print("Accuracy:", pipeline.score(X_test, y_test))
```
